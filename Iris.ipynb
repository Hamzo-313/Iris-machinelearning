```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
```

```python
Iris = load_iris()
```

```python
print(Iris.data)
print(Iris.data.shape)
print(Iris.target)
print(Iris.target.shape)
print(Iris.target_names)
```

```python
Iris_data = pd.DataFrame(Iris.data, columns=["がく片の長さ","がく片の幅","花びらの長さ","花びらの幅"])
Iris_target = pd.DataFrame(Iris.target , columns=["花の種類"])
Iris_all = pd.concat([Iris_data,Iris_target],axis=1)
display(Iris_data)
display(Iris_target)
display(Iris_all)
```

```python
input_train,input_test,output_train,output_test = train_test_split(Iris_data,Iris_target,test_size=0.33)
display(input_train)
display(output_train)
display(input_test)
display(output_test)
```

```python
plt.scatter(input_train[output_train["花の種類"]==0]["がく片の長さ"],input_train[(output_train==0).values]["がく片の幅"],label="setosa",cmap="rgb")
plt.scatter(input_train[output_train["花の種類"]==1]["がく片の長さ"],input_train[(output_train==1).values]["がく片の幅"],label="versicolor",cmap="rgb")
plt.scatter(input_train[output_train["花の種類"]==2]["がく片の長さ"],input_train[(output_train==2).values]["がく片の幅"],label="versicolor",cmap="rgb")
plt.xlim(3,9)
plt.ylim(1,5)
plt.xlabel("Length of sepal")
plt.ylabel("Width of sepal")
plt.title("Relation between length and width of sepal")
plt.legend()
```

```python
plt.scatter(input_train[output_train["花の種類"]==0]["花びらの長さ"],input_train[output_train["花の種類"]==0]["花びらの幅"]
           ,label='setosa',cmap="rgb")
plt.scatter(input_train[output_train["花の種類"]==1]["花びらの長さ"],input_train[output_train["花の種類"]==1]["花びらの幅"]
           ,label='versicolor',cmap="rgb")
plt.scatter(input_train[output_train["花の種類"]==2]["花びらの長さ"],input_train[output_train["花の種類"]==2]["花びらの幅"]
           ,label='virginica',cmap="rgb")
plt.xlim(0,8)
plt.ylim(0,3)
plt.xlabel("Length of petal")
plt.ylabel("Width of petal")
plt.title("Relationa between length and width of petal")
plt.legend()
```

```python
sepal_train = input_train[["がく片の長さ","がく片の幅"]]
petal_train = input_train[["花びらの長さ","花びらの幅"]]
display(sepal_train)
display(petal_train)
```

```python
#create each model
sepal_model = LinearSVC()
petal_model = LinearSVC()

#make each model training
sepal_model.fit(sepal_train,output_train)
petal_model.fit(petal_train,output_train)

#predict model by using test-data
sepal_predict = sepal_model.predict(input_test[["がく片の長さ","がく片の幅"]])
petal_predict = petal_model.predict(input_test[["花びらの長さ","花びらの幅"]])
```

```python
from sklearn.metrics import accuracy_score

sepal_score = accuracy_score(output_test,sepal_predict)
petal_score = accuracy_score(output_test,petal_predict)

print(f"Match rate between Length and Width of sepal: {sepal_score}")
print(f"Match rate between Length and Width of petal: {petal_score}")
```

```python
def Iris_border(H,M,model,param1,param2,resolution=0.01):
    H1_min,H1_max = H[param1].min()-0.5, H[param1].max()+0.5
    H2_min,H2_max = H[param2].min()-0.5, H[param2].max()+0.5
    H1,H2 = np.meshgrid(np.arange(H1_min,H1_max,resolution),
                           np.arange(H2_min,H2_max,resolution))
    n = np.array([H1.ravel(),H2.ravel()]).T
    Z = model.predict(n)
    Z = Z.reshape(H1.shape)
    plt.contourf(H1,H2,Z,alpha=0.5,cmap="Set2")
    plt.xlim(H1_min,H1_max)
    plt.ylim(H2_min,H2_max)
    plt.xlabel("Length")
    plt.ylabel("Width")
    plt.scatter(H[param1],H[param2],c=M["花の種類"],cmap="brg")
```

```python
Iris_border(input_train,output_train,sepal_model,"がく片の長さ","がく片の幅")
```

```python
Iris_border(input_train,output_train,petal_model,"花びらの長さ","花びらの幅")
```
